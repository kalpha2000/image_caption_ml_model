```python
def preprocess_img(img):
    img = image.load_img(img,target_size=(224,224))
    img = image.img_to_array(img)
    img = np.expand_dims(img,axis=0)
    # Normalisation
    img = preprocess_input(img)
    return img
```

```python
# this function extracts the feature after passsing through the restnet50
def encode_image(img):
    img = preprocess_img(img)
    feature_vector = model_new.predict(img)
    
    feature_vector = feature_vector.reshape((-1,))
#     print(feature_vector.shape)
    return feature_vector
```

# Data pre-processing for Captions

```python
# vocab creation with all the shortlisted words
len(total_words)
```

```python
word_to_idx = {}
idx_to_word = {}
# we will reserve the 0th pos in the dict for word that should not be there
# as the lstm will provide output, it can be that endseq is reached
# so for the rest we will se 0 position
for i,word in enumerate(total_words):
    word_to_idx[word] = i+1
    idx_to_word[i+1] = word
```

```python
print(len(idx_to_word),len(word_to_idx))
```

```python
# Two special words
idx_to_word[1846] = 'startseq'
word_to_idx['startseq'] = 1846

idx_to_word[1847] = 'endseq'
word_to_idx['endseq'] = 1847

# +1 because 0 idx is reserved
vocab_size = len(word_to_idx) + 1
print("Vocab Size",vocab_size)
```

```python
max_len = 0 
for key in train_descriptions.keys():
    for cap in train_descriptions[key]:
        max_len = max(max_len,len(cap.split()))
        
print(max_len)
```

# Data Loader (Generator)

```python
def data_generator(train_descriptions,encoding_train,word_to_idx,max_len,batch_size):
    # X1 is the img_feature vector S2 is the partial caption, y is next expected word in that caption
    X1,X2, y = [],[],[]
    
    n =0
    while True:
        for key,desc_list in train_descriptions.items():
            n += 1
            
            photo = encoding_train[key]
            for desc in desc_list:
                
                seq = [word_to_idx[word] for word in desc.split() if word in word_to_idx]
                for i in range(1,len(seq)):
                    # xi,yi, are single data points
                    # remember how xi is supposed to be 0,9,1,2,3,.....
                    # and then yi is supposed to denote the next word like 8
                    xi = seq[0:i]
                    yi = seq[i]
                    
                    #0 denote padding word
                    xi = pad_sequences([xi],maxlen=max_len,value=0,padding='post')[0]
                    yi = to_categorical([yi],num_classes=vocab_size)[0]
                    
                    X1.append(photo)
                    X2.append(xi)
                    y.append(yi)
                
                # for batch_size number of photos we yield as it is a generator function
                if n==batch_size:
                    yield ([np.array(X1),np.array(X2)],np.array(y))
                    X1,X2,y = [],[],[]
                    n = 0
```

# Word Embeddings

```python
f = open("./saved/glove.6B.50d.txt",encoding='utf8')
```

```python
embedding_index = {}
# embedding_index will have all embedding for all the words
for line in f:
    values = line.split()
    
    word = values[0]
    word_embedding = np.array(values[1:],dtype='float')
    embedding_index[word] = word_embedding
```

```python
f.close()
```

```python
embedding_index['apple']
```

```python
def get_embedding_matrix():
    emb_dim = 50
    matrix = np.zeros((vocab_size,emb_dim))
    for word,idx in word_to_idx.items():
        embedding_vector = embedding_index.get(word)
        
        if embedding_vector is not None:
            matrix[idx] = embedding_vector
            
    return matrix
```

```python
# embedding matrix will have embedding for the words which we require
# i.e, in the vocab
embedding_matrix = get_embedding_matrix()
embedding_matrix.shape
```

```python
# embedding matrix for 1846 and 1847 will be only 0 becauase that is reserved
# for empty and 1846 nd 1847 will be empty becasue they are for start and end
# symbols
```

# Model Architecture

```python
input_img_features = Input(shape=(2048,))
# inp_img1 = Dropout(0.3)(input_img_features)
inp_img2 = Dense(256,activation='relu')(input_img_features)
```

```python
# Captions as Input
input_captions = Input(shape=(max_len,))
inp_cap1 = Embedding(input_dim=vocab_size,output_dim=50,mask_zero=True)(input_captions)
# inp_cap2 = Dropout(0.3)(inp_cap1)
inp_cap3 = LSTM(256)(inp_cap1)
```

```python
decoder1 = add([inp_img2,inp_cap3])
decoder2 = Dense(256,activation='relu')(decoder1)
outputs = Dense(vocab_size,activation='softmax')(decoder2)

# Combined Model
model = Model(inputs=[input_img_features,input_captions],outputs=outputs)
```

```python
model.summary()
```

```python
# Important Thing - Embedding Layer
model.layers[2].set_weights([embedding_matrix])
model.layers[2].trainable = False
```

```python
model.compile(loss='categorical_crossentropy',optimizer="adam")
```

# Training of Model


```python
epochs = 20
batch_size = 3
steps = len(train_descriptions)//batch_size
```

```python
def train():
    
    for i in range(epochs):
        generator = data_generator(train_descriptions,encoding_train,word_to_idx,max_len,batch_size)
        model.fit(generator,epochs=1,steps_per_epoch=steps,verbose=1)
        model.save('./model_weights/model_'+str(i)+'.h5')
```

```python
train()
```

# Predictions

```python
def predict_caption(photo):
    
    in_text = "startseq"
    for i in range(max_len):
        sequence = [word_to_idx[w] for w in in_text.split() if w in word_to_idx]
        sequence = pad_sequences([sequence],maxlen=max_len,padding='post')
        
        ypred = model.predict([photo,sequence])
        ypred = ypred.argmax() #WOrd with max prob always - Greedy Sampling
        word = idx_to_word[ypred]
        in_text += (' ' + word)
        
        if word == "endseq":
            break
    
    final_caption = in_text.split()[1:-1]
    final_caption = ' '.join(final_caption)
    return final_caption
```
